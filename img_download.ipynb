```python
from bing_image_downloader import downloader

# List of categories to download
categories = ["penguin", "seal"]

# Loop through each category and download images
for category in categories:
    print(f"Downloading images for: {category}")
    downloader.download(
        category,
        limit=100,  # Change the limit as needed (e.g., 100 images per category)
        output_dir=f"images/{category}",
        adult_filter_off=True,
        force_replace=False,  # Set to True to overwrite existing images if needed
    )
```
